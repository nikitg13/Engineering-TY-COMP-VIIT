```
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns 
```

```
df = pd.read_csv("/content/data_housing.csv")
df.head()
```

```
df = df.astype({'bathrooms':int, 'floors':int})
df.head(3)
```

```
corrmat = df.corr()
fig = plt.figure(figsize=(12,9))
sns.heatmap(corrmat,vmax = 1,square = True)
plt.show()
```
